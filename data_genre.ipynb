```python
import numpy as np
import pandas as pd
import os
import re
```

```python
#load csv into a dataframe
df = pd.read_csv('scripts.csv')
df.head()
```

```python
#filter out the movies that contain action, horror or sci-fi as one of it's genres
genres_to_filter = {'action', 'horror', 'sci-fi'}

# Function to check if any genre of the movie is in the filter list
def is_genre_to_filter(genre_list):
    # Split the genre list and convert to lowercase, then check intersection
    movie_genres = set(g.strip().lower() for g in genre_list.split(','))
    return not genres_to_filter.isdisjoint(movie_genres)

# Apply the function and filter the DataFrame
df_filtered = df[~df['genres'].apply(is_genre_to_filter)]
```

```python
#save dataframe to csv
df_filtered.to_csv('filtered_scripts.csv', index=False)
```
